```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


# Taking input
light_name = input().strip()
action = input().strip().upper()

# Creating object
light = SmartLight(light_name)

# Performing action
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()

# Display output
light.display_status()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee Name:", self.name)
        print("Employee ID:", self.emp_id)
        print("Department:", self.department)


# Taking input
name = input().strip()
emp_id = input().strip()
department = input().strip()

# Creating object
employee = Employee(name, emp_id, department)

# Display ID card details
employee.display_id_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Taking input
name = input().strip()
phone = input().strip()

# Creating object
contact = Contact(name, phone)

# Display contact information
contact.display_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.name)
        print(f"Price: ₹{self.price}")


# Taking input
product_name = input().strip()
price = input().strip()

# Creating object
product = Product(product_name, price)

# Display price tag
product.display_price_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie Name:", self.name)
        print("Rating:", self.rating)


# Taking input
movie_name = input().strip()
rating = input().strip()

# Creating object
movie = Movie(movie_name, rating)

# Display movie details
movie.display_movie()
```

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Taking input
customer_name = input().strip()
address = input().strip()

# Creating object
delivery = Delivery(customer_name, address)

# Display delivery details
delivery.display_details()
```
